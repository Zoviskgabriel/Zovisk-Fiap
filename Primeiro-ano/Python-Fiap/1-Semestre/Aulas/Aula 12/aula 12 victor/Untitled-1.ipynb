```python
def faixa_int(minimo, maximo):
    while True:
        v = int(input("Entre com o valor entre {} e {}:".format(minimo, maximo)))
        if v < minimo or v  > maximo:
            print("Valor inválido. Digite um valor entre {} e {}".format(minimo, maximo))
        else:
            return v
        
minimo = 5
maximo = 10

valor = faixa_int(minimo, maximo)
print("Valor válido:", valor)
```

```python
def validar_string(minimo, maximo):
    while True:
        v = str(input("Escreva seu nome:"))

        if len(v ) < minimo or len(v)  > maximo:
            print("Nome inválido".format(minimo, maximo))
        else:
            return v

minimo = 5
maximo = 10

valor = validar_string(minimo, maximo)
print("O nome está correto:", valor)
```

```python
def verificar_string_na_lista(string, lista):
    return string in lista

nomes = ["João", "Maria", "José", "Ana"]
resultado = verificar_string_na_lista("José", nomes)
print(resultado)
```

```python
def soma(a, b):
    return a + b
def subtração(a, b):
    return a - b
def multi(a, b):
    return a * b
def divisao(a,b):
    return a/b

L_soma = [2, 3]
L_subtracao = [4,2]
L_multi = [3, 5]
L_div = [20, 4]
resultado_soma = soma(*L_soma)
resultado_sub = subtração(*L_subtracao)
resultado_multi = multi(*L_multi)
resultado_div = divisao(*L_div)

def calcular_media(resultado_soma, resultado_sub, resultado_multi, resultado_div):
    total = resultado_div + resultado_multi + resultado_soma + resultado_sub
    return total / 4

media = calcular_media(resultado_soma, resultado_sub, resultado_multi, resultado_div)

print("A média é:", media ) 
```
