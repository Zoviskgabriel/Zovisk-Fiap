```python
'''
- Função: Faixa_int
- Parãmetros: minimo e maximo 
-Entradas: V
- Condição: V < minimo ou V > maximo 
- Saida: valor invalido, retorna V
- como chamar função: faixa_int (minimo, maximo)
'''


def faixa_int(minimo, maximo):
    while True:
        v= int(input("Entre com um valor entre {} e {}:".format(minimo, maximo)))
        if v <minimo or v > maximo:
            print("Valor invalido. Digite um avalor entre {}e {}".format(minimo, maximo))
        else:
            return v
minimo = 5
maximo = 10

valor = faixa_int(minimo, maximo)
print("Valor invalido: ", valor)
```

```python
'''
Aplicação de função em dados (dataframe)
Somar valores de uma coluna numérica
'''

import pandas as pd

data = {'coluna_1': [10,20,30,40,50],
        'coluna_2': [5,10,15,20,25],
        'coluna_3': ['a','b','c','d','e'],
        'coluna_4': ['x','y','z','w','v']}

df = pd.DataFrame(data)

def soma_coluna_numerica(coluna):
    return coluna.sum()

resultado_soma = soma_coluna_numerica(df['coluna_1'])
resultado_soma
```
